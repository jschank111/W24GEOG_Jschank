measuring mean and median and comparing

```python
# showing mean and median
meanps = pparks['ACREAGE'].mean()
mean_meanps = round(meanps.mean(), 4)
print('The mean meanps in the data set is ', mean_meanps)
acreage = pparks['ACREAGE']
median_acreage = round(acreage.median(), 4)
print('The median acreage in the data set is ', median_acreage)
```

1. The mean and median are not the same. I can conclude from the mean and median that there are probably outliers in the data that are skewing the mean.
2. The distribution of Acreage is not evenly distributed

```python
# histogram

ax = pparks["ACREAGE"].hist(bins=20)
ax.set_xlabel('Philly Park Sizes in Acres')
ax.set_ylabel('Number of Parks')
ax.set_title('Histogram Showing the Distribution of Park Sizes in Philly')
ax.axvline(meanps, color='red', linestyle='solid', linewidth=2, alpha= .8)
ax.axvline(median_acreage, color='green', linestyle='solid', linewidth=2, alpha= .8)
ax.legend([meanps, median_acreage], labels = ['Mean', 'Median']);

```

This graph shows the distribution of Philly Park sizes. The mean and median show that the average and most common sizes of parks is close to 0.

```python
# calculating Standard Deviation
print('Std of variable acreage is ', round(pparks['ACREAGE'].std(), 4))
```

```python
# select only a subset of the parks: the pasks smaller than 10 acres
ax = pparks[~(pparks["ACREAGE"]=="Unknown")]["ACREAGE"].unique()
smallparks = pparks.loc[pparks['ACREAGE'] < 10]
smallparks.head()
```

```python
# calculating mean and median
mean_smallparks = round(smallparks ['ACREAGE'].mean(), 4)
print('The mean of smallparks in the data set is ', mean_smallparks)
median_smallparks = round(smallparks ['ACREAGE'].median(), 4)
print('The median of smallparks in the data set is ', median_smallparks)
```

```python
# histogram
ax = smallparks["ACREAGE"].hist(bins=20)
ax.set_xlabel('Philly Park Sizes in Acres')
ax.set_ylabel('Number of Parks')
ax.set_title('Histogram Showing the Distribution of Parks with sizes less than 10 acres in Philly')
ax.axvline(mean_smallparks, color='red', linestyle='solid', linewidth=2, alpha= .8)
ax.axvline(median_smallparks, color='green', linestyle='solid', linewidth=2, alpha= .8)
ax.legend([mean_smallparks, median_smallparks], labels = ['Mean', 'Median']);
```

This figure is showing the distribution of parks with less than 10 acres in philly. The reader should note that although all of these parks have less than 10 acres, the size of these parks fluctuates a lot from 0 to almost 10 acres.

```python
# Grouping the Parks based on the parent name
group_park = pparks.groupby(['PARENT_NAME']).sum()
group_park
```

```python
# plotting pairplot
pair = ['ACREAGE', 'Shape__Area']
pair
```

```python
sns.pairplot(group_park[pair], size=2.0)
```

There is a linear relationship between Acreage and Shape_Area. The larger the Acreage of the parent park the larger the Area, and visa versa.

```python
# calculating mean and median
groupmean = group_park['ACREAGE'].mean()
groupmedian = group_park['ACREAGE'].median()
print('The Mean of the grouped Parks is:', groupmean)
print('The Median of the grouped Parks is:', groupmedian)
```

```python
# Histogram of the parks grouped by Parent Name
ax = group_park["ACREAGE"].hist(bins=20)
ax.axvline(groupmean, color='red', linestyle='solid', linewidth=1.5, alpha = .8)
ax.axvline(groupmedian, color='green', linestyle='solid', linewidth=1.5, alpha = .8)
ax.legend([groupmean, groupmedian], labels = ['mean', 'median'])
ax.set_xlabel('Park size in acres')
ax.set_ylabel('Number of parks')
ax.set_title('Histogram Showing the Distribution of Parks Grouped by Parent Name')
```

9. This Graph is showing the number of parks grouped by parent name and their size in acres. It is also showing the mean and median of park size in acres which are both closer to 0 than 500. most parent parks have sizes of 50 acres or less.
10. This figure is almost identical to the first graph we made showing the distribution of park sizes in Philly. The mean and median are very similar too. I think this means that a lot of parks in Philly have a parent park, or are a parent park.

```python
# calculating standard deviation
print('Std of variable acreage is ', round(group_park['ACREAGE'].std(), 4))
```

11. This is a high standard deviation. It means that some of the data is very dispersed.
